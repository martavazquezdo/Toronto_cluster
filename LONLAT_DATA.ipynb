# Retrive latitude and longitude from every postal code

```python
import pandas as pd
```

```python
lonlat_data=pd.read_csv('http://cocl.us/Geospatial_data')

lonlat_data
```

## Sort the Canada table by Postal code to merge the latitude and longitude

Open the Canada data storged

```python
toronto_df = pd.read_csv("Canada_data.csv",sep=';')
```

Sort the table

```python
toronto_df=toronto_df.sort_values(by=['Postal Code'])
toronto_df = toronto_df.reset_index(drop=True)
toronto_df
```

Add latitude and longitude data

```python
toronto_df['Latitude']=lonlat_data.Latitude
toronto_df['Longitude']=lonlat_data.Longitude
toronto_df
```

Save the data for the exploration and clustering

```python
toronto_df.to_csv('Canada_lonlat.csv', sep=';', header=True, index=False)
```
